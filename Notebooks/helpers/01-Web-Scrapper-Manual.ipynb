# Função
- recebe a marca
- faz o request
- valida se tem algum link para consumir
- pega a lista de links
- itera sobre a lista de links para abrir a pagina X
- na página X pegar as infos:
    - nome
    - portas:
        - validar se é 2 ou 4
    - ano:
        - validar se tem 1 ano válido ou são 2
    - cambio:
        - validar se é automático ou manual
    - preço
- gravar em csv

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
def _crawler_list_link_cars(brand:str,page:str):
    """
        Função que recebe a marca para fazer um get no link mapeado
        Assim após o consumo, é necessário parsear com BeautifulSoup para que seja possível buscar todos os href
        Cada href será uma página de um produto específico
    """
    pag = page
    url_brand_test = f'https://www.icarros.com.br/comprar/usados/sp/{brand}?pag={pag}&ord=14'
    print(url_brand_test)
    url_brand_webcrawller = urlopen(url_brand_test)
    wmpage_brand_read =url_brand_webcrawller.read()
    page_soup = BeautifulSoup(wmpage_brand_read, "html.parser")
    list_links = [f'https://www.icarros.com.br{a["href"]}'  for a in page_soup.findAll("a", {"class":"clearfix"} , href=True)]
    return list_links
```

```python
def _validation_data_target(data_target_raw:object):
    data_target = {}
    title_validation = ''
    year_validation = ''
    color_validation = ''
    exchange_validation = ''
    port_validation = ''
    km_validation = ''
    
    if(data_target_raw['Nome']):
        title_validation =(data_target_raw['Nome']).strip()
        data_target['Nome'] = title_validation
    if(data_target_raw['Preco']):
        price_validation =(data_target_raw['Preco']).replace('R$','').strip().replace(',','')
        data_target['Preco'] = price_validation
    if(data_target_raw['Ano']):
        year_validation = (data_target_raw['Ano']).split()[0].strip()
        data_target['Ano'] = year_validation
    if(data_target_raw['Km']):
        km_validation = (data_target_raw['Km']).strip().replace('.','')
        data_target['Km'] = km_validation
    if(data_target_raw['Cor']):
        pass
    if(data_target_raw['Câmbio']):
        pass
    if(data_target_raw['Portas']):
        port = data_target_raw['Portas']
        port_validation = 4
        if(port.isnumeric()):
            port = int(port)
            if(port>2):
                port_validation = 4
        data_target['Portas'] = port_validation
        
    return data_target
```

```python
def _crawler_data(url:str):
    """
        Função que recebe uma url para pegar os dados alvo:
            - nome
            - preço
            - n portas
            - cor
            - kilometragem
            - ano
    """
    # requsição para dar get na página
    url_brand_webcrawller = urlopen(url)
    wmpage_brand_read =url_brand_webcrawller.read()
    # parseando
    page_soup = BeautifulSoup(wmpage_brand_read, "html.parser")
    
    # pegando dados da ul
    data_ul = page_soup.findAll("ul",{"class":"listahorizontal"})
    # buscando a ul com os dados corretos
    for ul in data_ul:
        list_data_span = ul.findAll("span",{"class":"destaque"})
        keys_target = ['Nome','Preco','Ano','Km','Câmbio','Cor']
        # validando se existe dados com o label keys_target
        if(len(list_data_span)>0):
            list_label = ul.findAll("h6")
            if len(list_label) >0:
                if(list_label[0].text in keys_target):
                    break

    # Pegando os labels
    list_label_data = [h.text for h in list_label]
    # Pegando os dados
    list_data = [data_target.text for data_target in list_data_span]
    
    
    #consolidação do objeto
    data_target = {}
    for i in range(len(list_label)):
        data_target[list_label_data[i]] = list_data[i]
        
    ## Pegando o Título
    data_target['Nome'] = (page_soup.find("h1",{"class":"titulo-sm"})).text
    ### Pegando o Preço
    data_target['Preco'] = (page_soup.find("h2",{"class":"preco"})).text
    
    #validação do objeto
    data_target_layout = {'Ano':'',
                          'Km': '',
                          'Cor': '',
                          'Câmbio': '',
                          'Portas': '',
                          'Preco':'',
                          'Nome':''}
    
    for key in data_target_layout.keys():
        if key not in data_target.keys():
            data_target[key] = ''
    
    #validação dados
    data_target_validate = _validation_data_target(data_target)
#     title_validation = ''
#     year_validation = ''
#     color_validation = ''
#     exchange_validation = ''
#     port_validation = ''
#     km_validation = ''
    
#     if(data_target['Nome']):
#         title_validation =(data_target['Nome']).strip()
#         data_target['Nome'] = title_validation
#     if(data_target['Preco']):
#         price_validation =(data_target['Preco']).replace('R$','').strip().replace(',','')
#         data_target['Preco'] = price_validation
#     if(data_target['Ano']):
#         year_validation = (data_target['Ano']).split()[0].strip()
#         data_target['Ano'] = year_validation
#     if(data_target['Km']):
#         km_validation = (data_target['Km']).strip().replace('.','')
#         data_target['Km'] = km_validation
#     if(data_target['Cor']):
#         pass
#     if(data_target['Câmbio']):
#         pass
#     if(data_target['Portas']):
#         port = data_target['Portas']
#         port_validation = 4
#         if(port.isnumeric()):
#             port = int(port)
#             if(port>2):
#                 port_validation = 4
#         data_target['Portas'] = port_validation
            
    
    

    
    return data_target_validate
```

```python
def web_crawler_ic(brand:str,page:str, n_product:str):
    lista_dados = []
    # loop para pegar numero de páginas necessárias
    for n_page in range(page):
        # pegando lista de produtos da página
        list_links_procucts = _crawler_list_link_cars(brand=brand,page=n_page)
        #loop para pegar cada link do produto
        if(n_product == 'max'):
            n_product = len(list_links_procucts)
        else:
            n_product = int(n_product)
        for n_link_product in range(n_product):
            data = _crawler_data(list_links_procucts[n_link_product])
            # pegando dados da página
            print(list_links_procucts[n_link_product])
            lista_dados.append(data)
    return lista_dados
    
```
